```python
"""This code calculates your grade per average to know if you pass or not your semester, then show graphics of your gpa"""

import matplotlib.pyplot as plt
import numpy as np
import tkinter as tk
from tkinter import ttk

coef_ue1={
    'R101':10,
    'R102':10,
    'R103':7,
    'R104':7,
    'R105':0,
    'R106':5,
    'R107':0,
    'R108':6,
    'R109':0,
    'R110':5,
    'R111':4,
    'R112':2,
    'R113':5,
    'R114':5,
    'R115':0,
    'SAE11':20,
    'SAE12':20,
    'SAE13':0,
    'SAE14':0,
    'SAE15':0,
    'SAE16':7,
    }

coef_ue2={
    'R101':4,
    'R102':0,
    'R103':2,
    'R104':8,
    'R105':6,
    'R106':0,
    'R107':0,
    'R108':0,
    'R109':0,
    'R110':5,
    'R111':5,
    'R112':2,
    'R113':9,
    'R114':9,
    'R115':3,
    'SAE11':0,
    'SAE12':0,
    'SAE13':29,
    'SAE14':0,
    'SAE15':0,
    'SAE16':7,
    }

coef_ue3={
    'R101':4,
    'R102':0,
    'R103':2,
    'R104':0,
    'R105':0,
    'R106':5,
    'R107':15,
    'R108':6,
    'R109':4,
    'R110':5,
    'R111':5,
    'R112':2,
    'R113':0,
    'R114':0,
    'R115':3,
    'SAE11':0,
    'SAE12':0,
    'SAE13':0,
    'SAE14':20,
    'SAE15':20,
    'SAE16':7,
    
}


#Display part
f = tk.Tk()

frame = ttk.Frame(f)
frame.grid()

f.title("Moyenne des UES")
f.geometry("500x700")

titre = ttk.Label(f, text="Moyennes:", font=("bold", 30))
titre.grid(column=0)

# Liste des matières
liste_matieres = [
    "R101","R102","R103","R104","R105","R106",
    "R107","R108","R109","R110","R111","R112",
    "R113","R114","R115",
    "SAE11","SAE12","SAE13","SAE14","SAE15","SAE16"
]

# Dictionnaire pour stocké les entrés de l'utilisateur
entries = {}

# Création automatique du texte et des zones pour écrire
for i, mat in enumerate(liste_matieres, start=2):
    lbl = ttk.Label(f, text=mat)
    lbl.grid(column=0, row=i)

    saisie = ttk.Entry(f)
    saisie.grid(column=1, row=i)
    entries[mat] = saisie  # on stocke chaque champ

# Fonction pour récupérer toutes les valeurs et créer le dictionnaire
def valider_toutes_les_notes():
    notes = {}
    for mat, entry in entries.items():
        try:
            notes[mat] = float(entry.get())
        except ValueError:
            notes[mat] = 0.0  # si rien ou invalide -> 0
    print(notes)  # affichage du dictionnaire dans la console
    return notes
    
notemoyenne=valider_toutes_les_notes()
print(notemoyenne)
# Bouton unique pour valider toutes les notes

btn_valider = ttk.Button(f, text="Valider toutes les notes", command=valider_toutes_les_notes)
btn_valider.grid(column=1, row=len(liste_matieres)+3, pady=20)


def calcul_ue(coef_ue:dict, notes:dict):
    """This function calculates the gpa of one UE"""
    #1.Initialisation
    gpa_ue=0
    total_coeff=0
    #2.Processing
    for subject in coef_ue:
        gpa_ue+=(notes[subject])*(coef_ue[subject])
        total_coeff+=coef_ue[subject]
    gpa_ue=(gpa_ue/total_coeff)
    #3.Conclusion
    return gpa_ue
    
#Tests

GPAS1UE1=calcul_ue(coef_ue1,notemoyenne)
GPAS1UE2=calcul_ue(coef_ue2,notemoyenne)
GPAS1UE3=calcul_ue(coef_ue3,notemoyenne)
print(GPAS1UE1)
print(GPAS1UE2)
print(GPAS1UE3)

def get_color(gpa):
    """Give a color according to your GPA"""
    if gpa >= 10: #above 10 color is green
        return 'green'
    elif 8 < gpa < 10: #between 8 and 10 color is orange
        return 'orange'
    else: #under 8 color is red 
        return 'red'
#Tests 

colorUE1=get_color(GPAS1UE1)
colorUE2=get_color(GPAS1UE2)
colorUE3=get_color(GPAS1UE3)
cs1=[colorUE1,colorUE2,colorUE3]

def pass_semester(cs:list):
    """This function help you know if you pass your semester"""
    #1.Initialisation
    validUE=0 #count your valid UE
    #2.Processing
    for color in cs: # browse around your colors of semester
        if color=='red':
            return False #One UE in red and you don't pass your semester
        if color=='green':
            validUE+=1
    #3.Conclusion
    if validUE>=2:
        return True
    else: 
        return False
#Tests

print(pass_semester(cs1))


plt.subplot(1,1,1)
plt.bar(['UE1','UE2','UE3'],[GPAS1UE1,GPAS1UE2,GPAS1UE3],color=cs1)
plt.title('GPA OF SEMESTER 1')
plt.show()
f.mainloop()
```
